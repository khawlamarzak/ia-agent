```python


import tiktoken

from dotenv.ipython  import load_dotenv
import os 
```

```python
load_dotenv(override = True )
```

```python
tokenizer = tiktoken.encoding_for_model("gpt-4o")
print(tokenizer.name)
```

```python
prompt = "vous etes un assistant expert dans le domaine de l analyse de sentiment"
tokens = tokenizer.encode(prompt)
print(tokens)
print("nombre de tokens : ", len(tokens))
```

```python
for token in tokens : 
    t = tokenizer.decode_single_token_bytes(token).decode('utf-8')
    print(t)
```

```python
def tokens_count(prompt : str , model : str = "o200k_base"):
    tokennizer = tiktoken.get_encoding(model)
    return len(tokennizer.encode(prompt))
```

```python
print(tokens_count(prompt = " vous etes un adulte "))
```
